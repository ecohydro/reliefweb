```python
def process_json(obj):  
    
    # Initialize lists to store the unpacked data
    dates = []
    countries = []
    countries_iso3 = []
    countries_ids = []
    primary_country = []
    primary_country_iso3 = []
    primary_country_id = []
    disaster_types = []
    disaster_types_ids = []
    theme_names = []
    theme_ids = []
    source_shortnames = []
    source_names = []
    titles = []
    urls = []
    disaster_names = []
    disaster_names_ids = []
    disaster_names_glides = []
    format_names = []
    format_ids = []
    language_names = []
    language_codes = []
    language_ids = []
    bodies = []

    # Iterate over the JSON data and unpack the elements
    for item in obj:
       # print(key, values)
        fields = item['fields']
        
        #date
        dates.append(fields['date']['original'])
        
        #country name
        if 'country' in fields and fields['country']:
            countries.append(', '.join([country['name'] for country in fields['country']]))
        else:
            countries.append(None)
        #country iso3
        if 'country' in fields and fields['country']:
            countries_iso3.append(', '.join([country['iso3'] for country in fields['country']]))
        else:
            countries_iso3.append(None)
        #country id
        if 'country' in fields and fields['country']:
            countries_ids.append(', '.join([str(country['id']) for country in fields['country']]))
        else:
            countries_ids.append(None)
        
        #primary country name
        if 'primary_country' in fields:
            primary_country.append(fields['primary_country']['name'])
        else:
            primary_countries.append(None)
        #primary_country iso3
        if 'primary_country' in fields:
            primary_country_iso3.append(fields['primary_country']['iso3'])
        else:
            primary_countries.append(None)
        #primary_country id
        if 'primary_country' in fields:
            primary_country_id.append(fields['primary_country']['id'])
        else:
            primary_country_id.append(None)

        #disaster type name
        if 'disaster_type' in fields and fields['disaster_type']:
            disaster_types.append(', '.join([disaster['name'] for disaster in fields['disaster_type']]))
        else:
            disaster_types.append(None)
        #disaster type id
        if 'disaster_type' in fields and fields['disaster_type']:
            disaster_types_ids.append(', '.join([str(disaster['id']) for disaster in fields['disaster_type']]))
        else:
            disaster_types_ids.append(None)    

        #theme name
        if 'theme' in fields and fields['theme']:
            theme_names.append(', '.join([theme['name'] for theme in fields['theme']]))
        else:
            theme_names.append(None)
        #theme id
        if 'theme' in fields and fields['theme']:
            theme_ids.append(', '.join([str(theme['id']) for theme in fields['theme']]))
        else:
            theme_ids.append(None)
        
        #disaster name
        if 'disaster' in fields and fields['disaster']:
            disaster_names.append(', '.join([disaster['name'] for disaster in fields['disaster']]))
        else:
            disaster_names.append(None)
        #disaster id
        if 'disaster' in fields and fields['disaster']:
            disaster_names_ids.append(', '.join([str(disaster['id']) for disaster in fields['disaster']]))
        else:
            disaster_names_ids.append(None)
        #disaster glide
        if 'disaster' in fields and fields['disaster']:
            if all('glide' in disaster for disaster in fields['disaster']):
                disaster_names_glides.append(', '.join(str(disaster['glide']) for disaster in fields['disaster']))
            else:
                disaster_names_glides.append(None)
        else:
            disaster_names_glides.append(None)
     
        #source short name
        #source_shortnames.append(fields['source'][0]['shortname'])
        if 'source' in fields and fields['source']:
            if all('shortname' in source for source in fields['source']):
                source_shortnames.append(', '.join(str(source['shortname']) for source in fields['source']))
            else:
                source_shortnames.append(None)
        else:
            source_shortnames.append(None)
        #source name
        #source_names.append(fields['source'][0]['name'])
        if 'source' in fields and fields['source']:
            if all('name' in source for source in fields['source']):
                source_names.append(', '.join(str(source['name']) for source in fields['source']))
            else:
                source_names.append(None)
        else:
            source_names.append(None)
        #format name
        #format_names.append(fields['format'][0]['name'])
        if 'format' in fields and fields['format']:
            if all('name' in source for source in fields['format']):
                format_names.append(', '.join(str(format_['name']) for format_ in fields['format']))
            else:
                format_names.append(None)
        else:
            format_names.append(None)
        #format id
       # format_ids.append(fields['format'][0]['id'])
        if 'format' in fields and fields['format']:
            if all('id' in source for source in fields['format']):
                format_ids.append(', '.join(str(format_['id']) for format_ in fields['format']))
            else:
                format_ids.append(None)
        else:
            format_ids.append(None)
        #title
        titles.append(fields['title'])
        #url
        urls.append(fields['url'])
        #language name
        #language_names.append(fields['language'][0]['name'])
        if 'language' in fields and fields['language']:
            if all('name' in source for source in fields['language']):
                language_names.append(', '.join(str(language['name']) for language in fields['language']))
            else:
                language_names.append(None)
        else:
            language_names.append(None)
        #language code
        #language_codes.append(fields['language'][0]['code'])
        if 'language' in fields and fields['language']:
            if all('code' in source for source in fields['language']):
                language_codes.append(', '.join(str(language['code']) for language in fields['language']))
            else:
                language_codes.append(None)
        else:
            language_codes.append(None)
        #language id
        #language_ids.append(fields['language'][0]['id'])
        if 'language' in fields and fields['language']:
            if all('id' in source for source in fields['language']):
                language_ids.append(', '.join(str(language['id']) for language in fields['language']))
            else:
                language_ids.append(None)
        else:
            language_ids.append(None)
            
        #body
        if 'body' in fields and fields['body']:
            bodies.append(fields['body'])
        else:
            bodies.append(None)
        
    # Create a DataFrame from the unpacked data
    data = {
        'date': dates,
        'countries': countries,
        'countries_iso3': countries_iso3,
        'countries_ids': countries_ids,
        'primary_country': primary_country,
        'primary_country_iso3': primary_country_iso3,
        'primary_country_id': primary_country_id,
        'disaster_types': disaster_types,
        'disaster_types_ids': disaster_types_ids,
        'disaster_names': disaster_names,
        'disaster_names_ids': disaster_names_ids,
        'disaster_names_glides': disaster_names_glides,
        'themes': theme_names,
        'theme_ids': theme_ids,
        'source_shortname': source_shortnames,
        'source_name': source_names,
        'format_name': format_names,
        'format_id': format_ids,
        'title': titles,
        'url': urls,
        'language_name': language_names,
        'language_codes': language_codes,
        'language_ids': language_ids,
        'body': bodies
    }
    
    df = pd.DataFrame(data)
    
    return df
```

```python
def extract_all_jsons(df):
    
    # List to store the processed DataFrames
    dfs = []

    # Iterate through the rows of the DataFrame
    for i, row in tqdm(df.iterrows()):
       # print(df[row])
        json_obj = row.to_dict()
        print(json_obj)
        # Process the JSON object and create a DataFrame
        #fields = row['fields']
        processed_df = process_json(json_obj)

        # Append the processed DataFrame to the list
        dfs.append(processed_df)

    # Concatenate the DataFrames into a single DataFrame
    result_df = pd.concat(dfs, ignore_index=True)
    
    return result_df

```

```python
def jsons_to_dfs(in_dir, out_dir):
    
    files=np.array(sorted(os.listdir(in_dir)))
    jsons = pd.Series(files).str.contains('.json')
    files = files[jsons]
    
    
    for filename in tqdm(files):
        base_name = os.path.splitext(filename)[0]
        
        with open(in_dir + filename, encoding = 'utf-8') as file:
            load = json.load(file)
            data = load['data']
            
        
        data_df = pd.DataFrame(data)
        
        
        extracted_data = process_json(data)
        #extracted_data.
    
        #extracted_data = extract_all_jsons(data_df)

        extracted_data.to_csv(out_dir + base_name + '.csv')

        del data
        del extracted_data
            
```

```python
jsons_to_dfs('data/disaster_jsons/', 'data/disaster_csvs/')
```

```python

# Directory containing the CSV files
csv_dir = 'data/disaster_csvs/'

# List to store the loaded DataFrames
dfs = []

# Iterate over the files in the directory
for filename in tqdm(os.listdir(csv_dir)):
    if filename.endswith('.csv'):
        # Read the CSV file into a DataFrame
        file_path = os.path.join(csv_dir, filename)
        df = pd.read_csv(file_path, index_col = [0])
        # Append the DataFrame to the list
        dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
concatenated_df = pd.concat(dfs)
concatenated_df = concatenated_df.reset_index(drop = True)

# Save the concatenated DataFrame as a new CSV file
output_file = 'data/disaster_reports.csv'
concatenated_df.to_csv(output_file)

print("CSV files concatenated and saved successfully.")
```
